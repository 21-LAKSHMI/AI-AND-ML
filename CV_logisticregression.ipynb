```python
#importing the Libraies
import numpy as np
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
# displaying first 5 rows
dataset.head()
```

```python
dataset.tail()
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
X=dataset[['Age', 'EstimatedSalary','Gender_Male']]
y=dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_predict
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.metrics import classification_report

# Example data: 3 input features
X, y = make_classification(n_samples=100, n_features=3, n_informative=3, n_redundant=0, random_state=42)

# Define minimal parameter grid
param_grid = {
    'C': [0.1, 1],         # Regularization strength
    'solver': ['liblinear'], # Solver that works with small datasets
    'penalty': ['l2']      # Regularization type
}

# GridSearchCV setup
grid = GridSearchCV(LogisticRegression(), param_grid, cv=3, verbose=1, n_jobs=-1)
grid.fit(X, y)

print("Best parameters:", grid.best_params_)
print("Best score:", grid.best_score_)

# Use cross_val_predict with 3-fold CV
y_pred = cross_val_predict(grid.best_estimator_, X, y, cv=3)

# Classification report
print(classification_report(y, y_pred))
```

```python
# Save Best Model
import pickle
filename = "CV_logisticregression.sav"
pickle.dump(grid.best_estimator_, open(filename, 'wb'))
```

```python
# CV Results table
results_df = pd.DataFrame(grid.cv_results_)
results_df.head()
```

```python
# Get input from the user
age = int(input("Enter Age: "))
estimated_salary = int(input("Enter Estimated Salary: "))
gender_male = int(input("Enter Gender (0 for Female, 1 for Male): "))

# Create a NumPy array with the user's input, ensuring the order of features matches the training data
user_input = np.array([[age, estimated_salary, gender_male]])

# Scale the user input using the same StandardScaler fitted on the training data
user_input_scaled = sc.transform(user_input)

# Load the saved model
loaded_model = pickle.load(open(filename, 'rb'))

# Make a prediction
prediction = loaded_model.predict(user_input_scaled)

# Display the output
if prediction[0] == 1:
    print("Prediction: Purchased (1)")
else:
    print("Prediction: Not Purchased (0)")
```
