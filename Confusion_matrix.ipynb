```python
# to analyze the data pandas is used
import pandas as pd
```

```python
# reading the dataset from folder
dataset=pd.read_csv("/content/college.csv")
```

```python
dataset
```

```python
# finding value counts for encourage
dataset[["Encourage"]].value_counts()
```

```python
# Importing onehot encoder
# changing the categorical values of gender and encourage to numerical column
from sklearn.preprocessing import OneHotEncoder
one=OneHotEncoder(handle_unknown='ignore')
dataset=pd.get_dummies(dataset,columns=['Gender','Encourage'],drop_first=True,dtype=int)
```

```python
dataset
```

```python
# splitting input and output values
x=dataset[["Parent_income",	"IQ",	"Gender_male",	"Encourage_not encourage"]]
y=dataset[["Plan"]]
```

```python
# importing train test split
from sklearn.model_selection import train_test_split
```

```python
# shuffling the dataset using random state
# 80% for training and 20% for testing dataset
X_train,X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
# importing decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# using fit function
reg=DecisionTreeClassifier()
reg.fit(X_train,y_train)
```

```python
# predicting the output
y_pred=reg.predict(X_test)
```

```python
# importing precision, recall and f1 score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
# caluculating precision, recall and f1 score
precision = precision_score(y_test, y_pred,average=None)
recall = recall_score(y_test, y_pred,average=None)
f1=f1_score(y_test, y_pred,average=None)
```
