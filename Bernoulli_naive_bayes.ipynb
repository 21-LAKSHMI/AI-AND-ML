```python
import pandas as pd
```

```python
# Read the preprocessed CSV file
dataset = pd.read_csv('/content/survey lung cancer.csv')
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.columns
```

```python
x=dataset[['SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
       'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH',
       'SWALLOWING DIFFICULTY', 'CHEST PAIN',]]
```

```python
y=dataset[['LUNG_CANCER']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.naive_bayes import BernoulliNB

# Initialize the Gaussian Naive Bayes model
model = BernoulliNB()

# Train the model using the training data
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# Calculate precision, recall, and F1-score (using 'weighted' average for multi-class classification)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-score: {f1}')
```

```python
import pickle
filename="Bernoulli.sav"
pickle.dump(model,open(filename,"wb"))
```

```python
# Get input from the user for each feature
input_values = {}
features = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']

for feature in features:
    while True:
        try:
            value = int(input(f"Enter the value for {feature} (1 or 2): "))
            if value in [1, 2]:
                input_values[feature] = value
                break
            else:
                print("Please enter either 1 or 2.")
        except ValueError:
            print("Invalid input. Please enter an integer.")

# Convert the dictionary to a pandas DataFrame with the correct column names
input_df = pd.DataFrame([input_values])

# Make the prediction using the loaded model
result = load_model.predict(input_df)

# Print the result
print(f'The predicted LUNG_CANCER result is: {result[0]}')
```
