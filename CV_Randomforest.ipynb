```python
#importing the Libraies
import numpy as np
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
X=dataset[['Age', 'EstimatedSalary','Gender_Male']]
y=dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_val_predict
from sklearn.metrics import classification_report

# Sample dataset with 3 input features (X1, X2, X3) and a target (y)
np.random.seed(42)
X = np.random.rand(100, 3)   # 100 samples, 3 features
y = np.random.choice([0, 1], size=100)  # Binary classification

# Define Random Forest model
rf = RandomForestClassifier(random_state=42)

# Minimal parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [50, 100],   # number of trees
    'max_depth': [None, 5]       # tree depth
}

# GridSearchCV with 3-fold cross-validation
grid = GridSearchCV(rf, param_grid, cv=3, n_jobs=-1, verbose=1)
grid.fit(X, y)

print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)

# Cross_val_predict with best model
y_pred = cross_val_predict(grid.best_estimator_, X, y, cv=3)

# Classification report
print("\nClassification Report:\n", classification_report(y, y_pred))
```

```python
# Save Best Model
import pickle
filename = "CV_randomforest.sav"
pickle.dump(grid.best_estimator_, open(filename, 'wb'))
```

```python
# CV Results table
results_df = pd.DataFrame(grid.cv_results_)
results_df.head()
```

```python
# Get user input for Age, EstimatedSalary, and Gender
age = int(input("Enter Age: "))
estimated_salary = int(input("Enter Estimated Salary: "))
gender = input("Enter Gender (Male/Female): ")

# Convert gender to numerical representation (1 for Male, 0 for Female)
gender_male = 1 if gender.lower() == 'male' else 0

# Create a numpy array from user input
user_input = np.array([[age, estimated_salary, gender_male]])

# Scale the user input using the same StandardScaler fitted on the training data
scaled_input = sc.transform(user_input)

# Make prediction using the best model
prediction = grid.best_estimator_.predict(scaled_input)

# Display the prediction.
if prediction[0] == 1:
    print("Prediction: Purchased")
else:
    print("Prediction: Not Purchased")
```
