```python
# to analyze the data pandas is used
import pandas as pd
```

```python
# reading the dataset from folder
dataset=pd.read_csv("/content/MBA.csv")
```

```python
dataset
```

```python
# counting the null values of all column
dataset.isnull().sum()
```

```python
# dropping race column, it has null values
dataset=dataset.drop(columns=['race'])
```

```python
dataset
```

```python
# counting the major column as groups
dataset[["major"]].value_counts()
```

```python
# counting the work_industry column as groups
dataset[["work_industry"]].value_counts()
```

```python
# importing Labelencoder from sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
# training the model using fit function
dataset['major']=le.fit_transform(dataset['major'])
dataset['work_industry']=le.fit_transform(dataset['work_industry'])
```

```python
# changed into label encoding
print(dataset[['major','work_industry']].head(10))
print(le.classes_)
```

```python
# splitting input and output values
x=dataset[['major','work_industry']]
y=dataset[['admission']]
```

```python
dataset
```

```python
# importing train test split
from sklearn.model_selection import train_test_split
```

```python
# shuffling the dataset using random state
# 80% for training and 20% for testing dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
# importing decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# using fit function
reg=DecisionTreeClassifier()
reg.fit(X_train,y_train)
```

```python
# predicting the output
y_pred=reg.predict(X_test)
```

```python
dataset
```

```python
# importing pickle
import pickle
```

```python
# save the file using sav extension
filename="DecisionTree.sav"
pickle.dump(reg,open(filename,"wb"))
```

```python
# load the saved model in pickle
load_model=pickle.load(open("/content/DecisionTree.sav","rb"))
```
