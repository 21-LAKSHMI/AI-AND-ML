```python
# importing pandas
# to analyze the data pandas is used
import pandas as pd
```

```python
# reading the dataset from the folder
dataset=pd.read_csv("/content/insurance_pre.csv")
```

```python
# displaying the dataset
dataset
```

```python
# converting the categorical column into numerical column
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
# displaying the dataset after encoding
dataset
```

```python
# separating x value
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
# separating y value
dep=dataset[['charges']]
```

```python
# importing train test split from sklearn
# 80% data for testing and 20% data for testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep, dep, test_size=0.20,random_state=0)
```

```python
# importing gridsearch cv from sklearn
# importing linear regression
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

# training the model using parameters

param_grid = {
    'fit_intercept': [True, False],
    'positive': [True, False]
}

# parameters are used to improve accuracy
grid = GridSearchCV(
    LinearRegression(),
    param_grid,
    refit=True,
    verbose=3,
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)

y_pred = grid.predict(X_test)
```

```python
# Assuming 'grid' is your GridSearchCV object and has been fitted to the data
print("Best parameters:", grid.best_params_)
# Access the mean test score for the best parameters
best_index = grid.best_index_
best_r2_score = grid.cv_results_['mean_test_score'][best_index]
# Printing the best R² score
print("r2:",best_r2_score)
```

```python
table=pd.DataFrame.from_dict(grid.cv_results_)
```

```python
table.sort_values('rank_test_score')
```

```python
# Assuming 'grid' is your GridSearchCV object that has been fitted to the data
best_model = grid.best_estimator_
```

```python
# importing pickle
# saving the file
import pickle
filename="RF_G.sav"
pickle.dump(best_model,open(filename,'wb'))
```
